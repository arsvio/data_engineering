 # Scaling: min-max normalization

```python
import numpy as np
from sklearn import preprocessing
```

Numpy is a package for array manipulation. Sklearn is a source providing ML packages. 

```python
# Define the features
fit_feat = fit.iloc[:,2:10]
fit_feat.head()
```

```python
# Min-max scaling
scaler = preprocessing.MinMaxScaler()
names = fit_feat.columns
aa = scaler.fit_transform(fit_feat)
scaled_df = pd.DataFrame(aa, columns=names)
scaled_df.head()
```

```python
aa
```

# Scaling to unit length

```python
x_array = np.array([3,2,7])
```

```python
normalized_arr = preprocessing.normalize([x_array])
print(normalized_arr)
```

```python
# One variable from fit, convert to array
active_min = np.array(fit['SedentaryMinutes'])
active_min[0:10]
```

```python
norm_min = preprocessing.normalize([active_min])
norm_min
```

```python
# More than one columns
fit_feat.head()
```

```python
names = fit_feat.columns
g = preprocessing.normalize(fit_feat)
scaled_df_norm = pd.DataFrame(g, columns=names)
scaled_df_norm.head()
```

Here the values are normalized along the rows, which can be very unintuitive. Normalizing along rows means that each individual sample is normalized instead of the features.

However, you can specify the axis while calling the method to normalize along a feature (column).

The value of axis parameter is set to 1 by default. If we change the value to 0, the process of normalization happens along a column.

```python
names = fit_feat.columns
g = preprocessing.normalize(fit_feat, axis=0)
scaled_df_norm = pd.DataFrame(g, columns=names)
scaled_df_norm.head()
```
